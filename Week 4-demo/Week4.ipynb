```python
##naming variables  

revenue = 125_000
expenses = 83_000
profit = revenue - expenses
profit
```

```python
taxes = 17_000
profit = revenue - expenses - taxes
print(profit)
```

```python
## add f-string with comma thousand separator
print(f"The profit after taxes is ${profit:,}")
```

```python
# change revenue, change expenses, change taxes for a loss
revenue = 85_000
expenses = 83_000
taxes = 17_000
loss = revenue - expenses - taxes
print(f"The loss after taxes is ${abs(loss):,}")
```
